```
# @title Define project information
PROJECT_ID = "ai-hangsik"  # @param {type:"string"}
LOCATION = "us-central1"  # @param {type:"string"}
```

```
# @title GCP Authentication

# Use OAuth to access the GCP environment.
import sys
if "google.colab" in sys.modules:
    from google.colab import auth
    auth.authenticate_user(project_id=PROJECT_ID)
```

```
# @title Enable Cloud translation API
!gcloud services enable translate.googleapis.com
```

```
# @title Import libraries and initialize vertex ai
import vertexai

vertexai.init(project=PROJECT_ID, location=LOCATION)
```

```
# @title Helper function for language detection

from google.cloud import translate

def detect_language() -> translate.DetectLanguageResponse:
    """Detecting the language of a text string.

    Args:

    Returns:
        The detected language of the text.
    """

    client = translate.TranslationServiceClient()
    parent = f"projects/{PROJECT_ID}/locations/{LOCATION}"

    # Detail on supported types can be found here:
    # https://cloud.google.com/translate/docs/supported-formats
    response = client.detect_language(
        content="안녕하세요!",
        parent=parent,
        mime_type="text/plain",  # mime types: text/plain, text/html
    )

    # Display list of detected languages sorted by detection confidence.
    # The most probable language is first.
    for language in response.languages:
        # The language detected
        print(f"Language code: {language.language_code}")
        # Confidence of detection result for this language
        print(f"Confidence: {language.confidence}")

    return response
```

```
# @title run
detect_language()
```
